```python
airlines = data['airline'].value_counts(ascending=True)
airlines.plot.barh(color=['lightgreen', 'lightblue'])
plt.title("Airlines with frequencies")
plt.xlabel("Number of flights")
plt.ylabel("Airline")
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index, data['departure_time'].value_counts().values, color= ('r', 'b'))
plt.title("Departure time")
plt.xlabel("D. Times")
plt.ylabel("D. Freq")
plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values, color= ('g', 'y'))
plt.title("Arrivval time")
plt.xlabel("V. Times")
plt.ylabel("V. Freq")
plt.show()
```

```python
source_city = data['source_city'].value_counts()
destination_city = data['destination_city'].value_counts()

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.barh(source_city.index,source_city.values, color = ('m', 'g'))
plt.title("Source Cities with No. of flights")
plt.xlabel("No. of flights")
plt.ylabel("Cities")
plt.subplot(1,2,2)
plt.barh(destination_city.index,destination_city.values, color = ('y', 'r'))
plt.title("Desstination Cities with No. of flights")
plt.xlabel("No. of flights")
plt.ylabel("Cities")
plt.show()
```
